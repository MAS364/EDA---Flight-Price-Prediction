```python
#Feature Engineering Process
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
#Drop Date of Journey Column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Departure_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.info
```

```python
df['Departure_Hour']=df['Departure_Hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
# Distribution of Flight Prices

plt.figure(figsize=(10, 6))
sn.histplot(df['Price'], bins=30, kde=True)
plt.title("Flight Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Number of Flight per Airline

plt.figure(figsize=(12, 6))
sn.countplot(y=df["Airline"], 
              order=df["Airline"].value_counts().index, 
              hue=df["Airline"],  # Assign 'Airline' to hue
              palette="coolwarm", 
              legend=False)  # Disable the legend
plt.title("Number of Flights per Airline")
plt.xlabel("Count")
plt.ylabel("Airline")
plt.show()
```

```python
# Impact of Stops on Flight Price

plt.figure(figsize=(8, 6))
sn.boxplot(x=df["Total_Stops"], y=df["Price"])
plt.title("Impact of Stops on Flight Price")
plt.xlabel("Total Stops")
plt.ylabel("Price")
plt.show()
```

```python

# Average Flight Price per Airline
plt.figure(figsize=(12, 6))
sn.barplot(x=df.groupby("Airline")["Price"].mean().index, 
            y=df.groupby("Airline")["Price"].mean().values, 
            hue=df.groupby("Airline")["Price"].mean().index,  # Assign hue to x
            palette="viridis", 
            legend=False)  # Disable the legend
plt.xticks(rotation=90)
plt.title("Average Flight Price per Airline")
plt.xlabel("Airline")
plt.ylabel("Average Price")
plt.show()
```
